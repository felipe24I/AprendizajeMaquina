1. Realizar la simulacin del ejercicio sobre el teorema de Bayes para las 2 cajas (roja y azul) y 2 frutas (manzana y naranja), donde se calcule:

- Las probabilidades posteriores
- Las verosimilitudes
- Las evidencias

IA utilizada (DeepSeek)

Prompt = "Hola deepSeek, quiero que realices una simulacin en python sobre un ejercicio del teorema de bayes:

Se tienen 2 eventos, cajas (B) y frutas (F), las cajas pueden tomar el valor rojo(r)(P(B=r)= 0.4) y azul(b)(P(B=b)= 0.6) las cuales representan las probabilidades previas, las frutas pueden tomar el valor manzana(a) o naranja(o), dentro de la caja roja  hay 6 naranjas y 2 manzanas, y dentro de la caja azul hay 1 naranja y 3 manzanas.

En base a la informacin anterior, y usando el teorema de bayes y la libreria NumPy ( solo si es necesario),  calcula todas las verosimilitudes,  las probabilidades posteriores y las evidencias, y adicional comenta el codigo"

```python

# Probabilidades previas de las cajas
P_Br = 0.4  # Probabilidad de caja roja
P_Bb = 0.6  # Probabilidad de caja azul

# Verosimilitudes (likelihoods) - probabilidad de fruta dado el tipo de caja
# Caja roja: 2 manzanas, 6 naranjas (total 8 frutas)
P_a_given_Br = 2/8  # P(manzana|caja roja)
P_o_given_Br = 6/8  # P(naranja|caja roja)

# Caja azul: 3 manzanas, 1 naranja (total 4 frutas)
P_a_given_Bb = 3/4  # P(manzana|caja azul)
P_o_given_Bb = 1/4  # P(naranja|caja azul)

# Calculamos la evidencia P(F) - probabilidad marginal de cada fruta
# P(a) = P(a|Br)P(Br) + P(a|Bb)P(Bb)
P_a = P_a_given_Br * P_Br + P_a_given_Bb * P_Bb
# P(o) = P(o|Br)P(Br) + P(o|Bb)P(Bb)
P_o = P_o_given_Br * P_Br + P_o_given_Bb * P_Bb

# Calculamos las probabilidades posteriores usando el Teorema de Bayes
# P(Br|a) = P(a|Br)*P(Br)/P(a)
P_Br_given_a = (P_a_given_Br * P_Br) / P_a
P_Bb_given_a = (P_a_given_Bb * P_Bb) / P_a

# P(Br|o) = P(o|Br)*P(Br)/P(o)
P_Br_given_o = (P_o_given_Br * P_Br) / P_o
P_Bb_given_o = (P_o_given_Bb * P_Bb) / P_o

# Mostramos resultados
print("=== Verosimilitudes ===")
print(f"P(a|Br) = {P_a_given_Br:.3f}")  # Probabilidad de manzana dado caja roja
print(f"P(o|Br) = {P_o_given_Br:.3f}")  # Probabilidad de naranja dado caja roja
print(f"P(a|Bb) = {P_a_given_Bb:.3f}")  # Probabilidad de manzana dado caja azul
print(f"P(o|Bb) = {P_o_given_Bb:.3f}")  # Probabilidad de naranja dado caja azul

print("\n=== Evidencias (Probabilidades marginales) ===")
print(f"P(a) = {P_a:.3f}")  # Probabilidad total de manzana
print(f"P(o) = {P_o:.3f}")  # Probabilidad total de naranja

print("\n=== Probabilidades posteriores ===")
print(f"P(Br|a) = {P_Br_given_a:.3f}")  # Probabilidad de caja roja dado que se sac manzana
print(f"P(Bb|a) = {P_Bb_given_a:.3f}")  # Probabilidad de caja azul dado que se sac manzana
print(f"P(Br|o) = {P_Br_given_o:.3f}")  # Probabilidad de caja roja dado que se sac naranja
print(f"P(Bb|o) = {P_Bb_given_o:.3f}")  # Probabilidad de caja azul dado que se sac naranja
```

2. Realizar la simulacin del ejercicio sobre el teorema de Bayes para n cajas y m frutas, donde se calcule:

- Las probabilidades posteriores
- Las verosimilitudes
- Las evidencias

IA utilizada (DeepSeek)

Prompt = "Super bien, ese es el resultado que esperaba. Ahora realiza la simulacin donde:

Los eventos sean n cajas y m frutas que van a ser ingresadas por el usuario (con n y m >0, si alguna no cumple, ahi mismo arrojar mensaje de error y volver a preguntarle), le vas a pedir el color de la caja y el nombre de la fruta, y la probabilidad previa (debe estar entre 0 y 1, sino ocurre algo de eso, sacar mensaje de error y vuelvele a preguntar al usuario. Aparte la suma debe dar 1, por tanto en la probabilidad previa de la ultima caja, asignarla automatico para que la suma de 1 y escribir en mensaje avisando eso, asi no le vas a preguntar la probabilidad previa de la ultima caja al usuario), despues le vas a pegruntar por cada caja la cantidad de frutas que hay en esa caja. Finalmente vas a calcular los mismo parametros del anterior prompt, adems comenta el cdigo dentro del programa"

```python
import numpy as np

def obtener_entero_positivo(mensaje):
    """Solicita un entero positivo al usuario con validacin"""
    while True:
        try:
            valor = int(input(mensaje))
            if valor > 0:
                return valor
            else:
                print("Error: El valor debe ser mayor que 0. Intente nuevamente.")
        except ValueError:
            print("Error: Debe ingresar un nmero entero. Intente nuevamente.")

def obtener_probabilidad(mensaje, suma_actual=0):
    """
    Solicita una probabilidad vlida al usuario (entre 0 y 1)
    y que no haga que la suma acumulada exceda 1
    """
    while True:
        try:
            valor = float(input(mensaje))
            if 0 <= valor <= 1:
                if suma_actual + valor <= 1:
                    return valor
                else:
                    print(f"Error: La suma acumulada excedera 1. Mximo permitido: {1-suma_actual:.2f}")
            else:
                print("Error: La probabilidad debe estar entre 0 y 1.")
        except ValueError:
            print("Error: Debe ingresar un nmero. Intente nuevamente.")

def main():
    print("=== Simulacin del Teorema de Bayes ===")
    print("Este programa calcula probabilidades posteriores dados tipos de cajas y frutas.")

    # Paso 1: Obtener nmero de cajas y frutas
    n = obtener_entero_positivo("Ingrese el nmero de cajas (n > 0): ")
    m = obtener_entero_positivo("Ingrese el nmero de tipos de frutas (m > 0): ")

    # Paso 2: Obtener informacin de las cajas
    cajas = []
    frutas = []
    suma_prob = 0.0

    # Primero obtener los nombres de las frutas
    print("\n=== Ingreso de tipos de frutas ===")
    for i in range(m):
        fruta = input(f"Ingrese el nombre de la fruta {i+1}: ").strip()
        frutas.append(fruta)

    # Luego obtener informacin de cada caja
    print("\n=== Ingreso de informacin de cajas ===")
    for i in range(n):
        print(f"\nCaja {i+1}:")
        color = input("Ingrese el color de la caja: ").strip()

        # Manejar probabilidad previa (la ltima se calcula automticamente)
        if i < n-1:
            prob = obtener_probabilidad(
                f"Ingrese la probabilidad previa para la caja {color} (0-1, suma acumulada: {suma_prob:.2f}): ",
                suma_prob)
            suma_prob += prob
        else:
            prob = 1 - suma_prob
            print(f"Probabilidad automtica para la ltima caja {color}: {prob:.2f} "
                  f"(para completar suma=1.00)")

        # Obtener distribucin de frutas en esta caja
        dist_frutas = {}
        total_frutas = 0
        print(f"\nDistribucin de frutas para la caja {color}:")

        for j, fruta in enumerate(frutas):
            while True:
                try:
                    cantidad = int(input(f"Ingrese cantidad de {fruta} en caja {color}: "))
                    if cantidad >= 0:
                        dist_frutas[fruta] = cantidad
                        total_frutas += cantidad
                        break
                    else:
                        print("Error: La cantidad no puede ser negativa.")
                except ValueError:
                    print("Error: Debe ingresar un nmero entero.")

        if total_frutas == 0:
            print("Advertencia: La caja no tiene frutas. Se asignar distribucin uniforme.")
            for fruta in frutas:
                dist_frutas[fruta] = 1
            total_frutas = m

        # Almacenar informacin de la caja
        cajas.append({
            'color': color,
            'prob_prev': prob,
            'dist_frutas': dist_frutas,
            'total_frutas': total_frutas
        })

    # Paso 3: Calcular verosimilitudes (likelihoods)
    print("\n=== Clculo de verosimilitudes ===")
    likelihoods = {}
    for caja in cajas:
        color = caja['color']
        likelihoods[color] = {}
        for fruta in frutas:
            # P(fruta|caja) = cantidad de esa fruta / total frutas en caja
            prob = caja['dist_frutas'].get(fruta, 0) / caja['total_frutas']
            likelihoods[color][fruta] = prob
            print(f"P({fruta}|{color}) = {prob:.3f}")

    # Paso 4: Calcular probabilidades marginales (evidencia)
    print("\n=== Clculo de evidencias (probabilidades marginales) ===")
    marginal_probs = {}
    for fruta in frutas:
        # P(fruta) = sumatoria sobre todas las cajas de P(fruta|caja)*P(caja)
        prob = sum(caja['prob_prev'] * likelihoods[caja['color']][fruta] for caja in cajas)
        marginal_probs[fruta] = prob
        print(f"P({fruta}) = {prob:.3f}")

    # Paso 5: Calcular probabilidades posteriores
    print("\n=== Clculo de probabilidades posteriores ===")
    posterior_probs = {}
    for caja in cajas:
        color = caja['color']
        posterior_probs[color] = {}
        for fruta in frutas:
            # P(caja|fruta) = P(fruta|caja)*P(caja)/P(fruta)
            if marginal_probs[fruta] > 0:
                prob = (likelihoods[color][fruta] * caja['prob_prev']) / marginal_probs[fruta]
            else:
                prob = 0.0  # Evitar divisin por cero
            posterior_probs[color][fruta] = prob
            print(f"P({color}|{fruta}) = {prob:.3f}")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

class GenerateData:
    def __init__(self, N=100, seed=None):
        self.N = N
        if seed is not None:
            np.random.seed(seed)

        self.params = {
            1: {"mean": 0, "std": 1},
            2: {"mean": 3, "std": 1.5}
        }

        self.X = None
        self.Y = None

    def generate(self):
        probs = np.random.dirichlet(alpha=[1,1])
        self.Y = np.random.choice([1, 2], size=self.N, p=probs)
        self.X = np.array([
            np.random.normal(
                loc=self.params[y]["mean"],
                scale=self.params[y]["std"]
            ) for y in self.Y
        ])

    def plot_all(self):
        fig, axs = plt.subplots(2, 2, figsize=(12, 8))

        # --- Subplot (0,0): Dispersin p(X,Y)
        axs[0, 0].scatter(self.X, self.Y, color='blue', s=25)
        axs[0, 0].set_title(r'$p(X,Y)$')
        axs[0, 0].set_xlabel(r'$X$')
        axs[0, 0].set_ylabel(r'$Y$')
        axs[0, 0].set_yticks([1, 2])
        axs[0, 0].grid(False)

        # Lneas rojas tipo cuadrcula
        x_min, x_max = self.X.min(), self.X.max()
        x_lines = np.linspace(x_min, x_max, 9)
        for xc in x_lines:
            axs[0, 0].axvline(x=xc, color='red', lw=2)
        axs[0, 0].axhline(y=1.5, color='red', lw=2)

        # --- Subplot (0,1): Histograma p(Y) (barras horizontales corregido)
        counts = [(self.Y == y).sum() for y in [2, 1]]  # primero 2, luego 1
        probs = np.array(counts) / self.N
        axs[0, 1].barh([2, 1], probs, color='lavender', edgecolor='black')
        axs[0, 1].set_title(r'$p(Y)$')
        axs[0, 1].set_yticks([1, 2])
        axs[0, 1].set_ylim(0.5, 2.5)  # para que queden centrados
        axs[0, 1].set_xlim(0, max(probs)*1.2)

        # --- Subplot (1,0): Histograma p(X)
        axs[1, 0].hist(self.X, bins=9, color='lavender', edgecolor='black', density=True)
        axs[1, 0].set_title(r'$p(X)$')
        axs[1, 0].set_xlabel(r'$X$')

        # --- Subplot (1,1): Histograma p(X|Y=1)
        x_given_1 = self.X[self.Y == 1]
        axs[1, 1].hist(x_given_1, bins=9, color='lavender', edgecolor='black', density=True)
        axs[1, 1].set_title(r'$p(X|Y=1)$')
        axs[1, 1].set_xlabel(r'$X$')

        plt.tight_layout()
        plt.show()


def main():
  # Crear objeto
  data = GenerateData(N=100, seed=42)

  # Generar datos
  data.generate()

  # Graficar todo en una figura 2x2
  data.plot_all()

if __name__ == "__main__":
  main()
```
